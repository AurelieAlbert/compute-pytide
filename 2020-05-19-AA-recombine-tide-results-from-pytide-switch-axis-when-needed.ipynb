```python
import netCDF4
import xarray as xr
import os
import glob
import numpy as np

%matplotlib inline
%autosave 60
```

```python
def combine(k):
    print('combining for time-step '+str(k))
    tide_x1=xr.open_dataset('/work/ALT/odatis/eNATL60/outputs/pytide/eNATL60miX1-BLBT02_tide_y2009m07-y2010m06_k'+str(k)+'.nc')
    tide_x2=xr.open_dataset('/work/ALT/odatis/eNATL60/outputs/pytide/eNATL60miX2-BLBT02_tide_y2009m07-y2010m06_k'+str(k)+'.nc')
    if len(tide_x1.x) != 4177:
        tide_x1_rename=tide_x1.rename({'x':'Y','y':'X'})
        tide_x1_concat=tide_x1_rename.rename({'Y':'y','X':'x'})
    else:
        tide_x1_concat=tide_x1
    if len(tide_x2.x) != 4177:
        tide_x2_rename=tide_x2.rename({'x':'Y','y':'X'})
        tide_x2_concat=tide_x2_rename.rename({'Y':'y','X':'x'})
    else:
        tide_x2_concat=tide_x2
    tide=xr.concat([tide_x1_concat,tide_x2_concat], dim='x')
    tide.to_netcdf(path='/work/ALT/odatis/eNATL60/outputs/pytide/eNATL60-BLBT02_tide_y2009m07-y2010m06_k'+str(k)+'.nc',engine='scipy')
```

```python
for k in np.arange(2571,8735):
    combine(k)
```
